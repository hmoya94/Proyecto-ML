```python
import pandas as pd

train_df = pd.read_csv('../data/raw/train.csv')
test_df = pd.read_csv('../data/raw/test.csv')
```

```python
train_df = train_df.drop('id', axis=1)
```

```python
fumadores = train_df[train_df['smoking'] == 1]
no_fumadores = train_df[train_df['smoking'] == 0]
```

```python
stats_fumadores = fumadores.describe()
stats_no_fumadores = no_fumadores.describe()

stats_fumadores
```

```python
stats_no_fumadores
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

features = ['age', 'weight(kg)', 'height(cm)', 'waist(cm)']
for feature in features:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='smoking', y=feature, data=train_df)
    plt.title(f'Distribución de {feature} por Estado de Fumador')
    plt.show()
```

```python
# Por ejemplo, para una característica categórica ficticia llamada 'category_feature'
pd.crosstab(train_df['smoking'], train_df['category_feature'], normalize='index')
```

```python
correlaciones = train_df.corr()['smoking'].sort_values()
correlaciones
```

```python
from sklearn.linear_model import LogisticRegression

modelo = LogisticRegression()
modelo.fit(train_df[features], train_df['smoking'])
importancia = pd.DataFrame(modelo.coef_[0], index=features, columns=['importancia']).sort_values('importancia', ascending=False)
importancia
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

corr = fumadores.corr()
plt.figure(figsize=(22,18))  
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Supongamos que tienes dos DataFrames: fumadores y no_fumadores
# fumadores = pd.DataFrame(...) # Este es tu DataFrame de fumadores
# no_fumadores = pd.DataFrame(...) # Este es tu DataFrame de no fumadores

# Para comparar, por ejemplo, la edad, peso y circunferencia de la cintura entre fumadores y no fumadores
variables_a_comparar = fumadores.columns.tolist()

for variable in variables_a_comparar:
    plt.figure(figsize=(10, 6))
    sns.histplot(fumadores[variable], color="red", label="Fumadores", kde=True, stat="density", linewidth=0)
    sns.histplot(no_fumadores[variable], color="blue", label="No Fumadores", kde=True, stat="density", linewidth=0)
    plt.legend()
    plt.title(f'Comparación de {variable}')
    plt.xlabel(variable)
    plt.ylabel('Densidad')
    plt.show()
```

```python
if 'smoking' in variables_a_comparar:
    variables_a_comparar.remove('smoking') # Excluye 'smoking' si es una columna numérica

sns.set(style="whitegrid")
plt.figure(figsize=(15, 10)) 

for i, variable in enumerate(variables_a_comparar, 1):
    plt.subplot(len(variables_a_comparar) // 3 + 1, 3, i) 
    sns.kdeplot(fumadores[variable], color="red", label="Fumadores", fill=True)
    sns.kdeplot(no_fumadores[variable], color="blue", label="No Fumadores", fill=True)
    plt.title(variable)
    plt.xlabel('')
    plt.tight_layout()

plt.show()
```
